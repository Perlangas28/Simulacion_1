**Prueba aplicada:** Jackknife

Sea una muestra de $n$ datos. En esta prueba eliminamos uno de los datos para obtener $n$ cantidad de muestras de tamaño $n-1$.

Para cada muestra calcularemos el estadístico $\hat{\theta}_i$ para después obtener el promedio de Jackknife:

$$\theta=\frac{1}{n}\sum_{i=1}^1{n}\hat{\theta}_i$$
y la varianza:
$$\mathrm{Var}_{Jack}=\frac{n-1}{n}\sum_{i=1}^{n}\left(\theta-\hat{\theta}_i\right)^2$$

```python
import numpy as np
import random as ran
import statistics as st
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
datos  = [35, 42, 38, 40, 45, 37, 39, 41, 44, 36, 43, 40]

n = len(datos)

def jackkinfe(datos):
  n = len(datos)
  datos = np.array(datos)
  theta_i = np.empty(n)
  for i in range(n):
    muestra = np.delete(datos, i)
    theta_i[i] = muestra.mean()
  return theta_i

theta_i = jackkinfe(datos)

theta_jack = theta_i.mean()
var_jack = (n-1)/n * np.sum((theta_jack - theta_i)**2)
ee_jack = np.sqrt(var_jack)

print("Media muestral original:", st.mean(datos))
print("Media muestral Jackknife:", theta_jack)
print("Varianza Jackknife:", var_jack)
print("Error estándar Jackknife:", ee_jack)
```

```python
promedio_jack = sum(theta_i)/len(theta_i)

plt.plot(theta_i, marker='o')
plt.axhline(y=promedio_jack, color='r', linestyle='--', label='Promedio Jackknife')
plt.show()
```
