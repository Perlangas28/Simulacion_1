**Estime**

$∫_0^1 exp(x^2)\,dx$

**Por método Monte Carlo.**

**Genere al menos $100$ valores y deténgase cuando la desviación estándar de su estimador sea menor que $0.01$, $0.001$.**


```python
import numpy as np
```

Definimos la precisión deseada, es decir, el valor máximo permitido para el error estándar del estimador. El programa se detendrá cuando:

$S/\sqrt{k} < d$

Donde $S$ es la desviación estándar muestral y $k$ es el número de simulaciones realizadas.

```python
d = 0.01
```

Generamos el primer número aleatorio, un número uniforme entre $0$ y $1$.

Después evaluamos la función.

```python
u = np.random.uniform(0,1)
x = np.exp(u**2)
```

Inicializamos la media, la varianza y el contador.

```python
media = x
S2 = 0
j = 1
```

Hacemos un bucle para seguir simulando hasta tener al menos $100$ observaciones y hasta que el error estándar $S/\sqrt{k}<0.01$.
Finalmente la media obtenida es la estimación Monte Carlo de la integral.

```python

while j < 1000 or np.sqrt(S2)/np.sqrt(j) >= d:

  nuevo_u = np.random.uniform(0,1)
  x_nuevo = np.exp(nuevo_u**2)

  media_anterior = media

  media = media + (x_nuevo - media)/(j+1)

  S2 = (1 - 1/j)*S2 + (j+1)*(media - media_anterior)**2

  j += 1

error_estandar = np.sqrt(S2)/np.sqrt(j)

print("Número de simulaciones:", j)
print("Estimación:", media)
print("Desviación estándar:", np.sqrt(S2))
print("Error estándar:", error_estandar)
```

En conclusión, al realizar el método por simulación y de manera analítica, se obtiene un resultado con una diferencia mínima de $0.00163$. Por lo tanto el método Monte Carlo es capaz de aproximar la integral cuando se utiliza un número suficiente de simulaciones.
