```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def uniforme(a,b,n):
  l = []
  for i in range(n):
    u = r.random()
    x = (b-a)*u+a
    l.append(x)
  return l
```

```python
M=uniforme(2,5,1000)
```

```python
plt.hist(M,bins=50)
```

```python
def cdf(M):
  n=len(M)
  x=np.sort(M)
  y=np.arange(0,n)/n
  return x,y
```

```python
x_math,y_math=cdf(M)
```
