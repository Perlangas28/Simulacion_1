```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def gaussiana(mu, sigma, n):
    muestra = []
    for i in range(n):
        suma = 0
        for j in range(12):
            u = r.random()
            suma = suma + u
        x = mu + sigma * (suma - 6)
        muestra.append(x)
    return muestra
```

Los datos del problema son:

```python
error = 1
n = 100000

mu1 = 1.5
var1 = 0.0016
sigma1 = np.sqrt(var1)
mu2 = 1.48
var2 = 0.0009
sigma2 = np.sqrt(var2)
p_analitica = 0.3446
```

```python
X1 = gaussiana(mu1, sigma1, n)
X2 = gaussiana(mu2, sigma2, n)

interferencia = 0
for i in range(n):
  if X2[i] > X1[i]:
    interferencia += 1
```

```python
p_simulada = interferencia / n

print("La probabilidad simulada es:", p_simulada)
```

```python
error = abs(p_simulada - p_analitica)
if error > 0.01:
  n +=100
print("simulaciones", n)
print("error", error)
```
