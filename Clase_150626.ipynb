```python
import numpy as np
import matplotlib.pyplot as plt
import random as r
```

```python
def f(x):
  return 1/(1+x)
```

```python
f(0.5)
```

```python
def mc1(f,N):
  s = 0
  for i in range (N):
    U = r.random()
    s += f(U)
  return s/N
```

```python
l = []
for i in range (100):
  M = mc1(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
def mc2(f,N):
  s = 0
  for i in range (N):
    U = r.random()
    s += f(1-U)
  return s/N
```

```python
l = []
for i in range (100):
  M = mc2(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
def mc(f,N):
  s = 0
  for i in range (N):
    U = r.random()
    s += (f(U)+f(1-U))/2
  return s/N
```

```python
l = []
for i in range (100):
  M = mc(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```
