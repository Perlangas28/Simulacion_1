```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def Poisson(lam,T):
  t=0
  S=[]
  I=0
  P=[]
  while t<=T:
    R=r.random()
    t=t-((1/lam)*np.log(R))
    I=I+1
    S.append(t)
    P.append(I)
  return S,P
```

```python
L=Poisson(4,10)
```
