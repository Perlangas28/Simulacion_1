La distribución de Poisson es una distribución de probabilidad discreta que se utiliza para modelar la cantidad de veces que ocurre un evento en un intervalo fijo de tiempo, espacio o cualquier otra unidad, cuando los eventos ocurren de manera independiente y con una tasa promedio constante.
La función de probabilidad de la distribución de Poisson es:

$P(X = x) = e^{−λ}λ^x / {x!}$
	​

Donde:

X = número de eventos.

λ = media o tasa promedio de ocurrencia.

e = constante de Euler.

x = número de eventos observados.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

Primero se define el valor de λ, que representa el promedio de eventos esperados y se generan distintos valores de X (cantidad de eventos).

```python
lambda_poisson = 4
x = np.arange(0, 15)
```

Después se calcula la probabilidad para cada valor de X.

```python
probabilidades = poisson.pmf(x, lambda_poisson)
print("Distribución de Poisson")
print(f"Lambda = {lambda_poisson}\n")

for i in range(len(x)):
    print(f"P(X = {x[i]}) = {probabilidades[i]:.5f}")
```

Finalmente, se imprimen los resultados y se genera una gráfica de barras para visualizar la distribución.

```python
plt.figure(figsize=(8,5))
plt.bar(x, probabilidades)

plt.show()
```

Si λ=4, significa que en promedio ocurren 4 eventos por intervalo. El programa calcula la probabilidad de que ocurran exactamente 0, 1, 2, 3, ..., 14 eventos.

La gráfica permite observar cuáles cantidades de eventos son más probables y cómo las probabilidades disminuyen conforme los valores se alejan de la media.
